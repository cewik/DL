## Preprocessing of Data
- Train | Test Split, Scalling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
#import pickle
#pickle.dump(scaler, open("scaler_kc_house", 'wb'))
```

```python
df = pd.get_dummies(df)
df.head()
```

```python
X = df.drop('Exited', axis=1)
y = df['Exited'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, test_size = 0.10, random_state = 42)
```

```python
scaler = MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
def shape_control():
    print('df.shape:', df.shape)
    print('X.shape:', X.shape)
    print('y.shape:', y.shape)
    print('X_train.shape:', X_train.shape)
    print('y_train.shape:', y_train.shape)
    print('X_test.shape:', X_test.shape)
    print('y_test.shape:', y_test.shape)
shape_control()
```

## Modelling & Model Performance

### Import related libraries

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import plot_roc_curve, roc_auc_score, roc_curve
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.model_selection import GridSearchCV
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
# tenserflow sequential yapsn sarp bir nesne haline getirip cv ye veriyoruz.
```

```python
seed = 42
```

### Creating Model

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
tf.random.set_seed(seed) 

model = Sequential()
model.add(Dense(32, activation = "relu", input_dim = X_train.shape[1]))  # sigunt ve hiperb. e gre renme daha fazla olduu iin relu kullandk.
model.add(Dense(16, activation = "relu"))
model.add(Dense(1, activation = "sigmoid"))  # output binary olduu (iin 1neuron var) sigmoid kullandk. 

model.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
early_stop = EarlyStopping(monitor = "val_loss", mode = "auto", verbose = 1, patience = 30)
```

### without class_weigth

```python
model.fit(x = X_train, y = y_train, validation_split = 0.1, batch_size = 32, epochs = 1000, verbose = 1,
          callbacks = [early_stop])
```

#### Evaluate

```python
model.summary()
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.head()
```

```python
loss_df.plot();
```

```python
model.evaluate(X_test, y_test, verbose=0)
```

```python
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print("loss : ", loss)
print("accuracy : ", accuracy)
```

```python
y_pred = (model.predict(X_test) > 0.5).astype("int32")   # 1 olma ihtimali 0.5 ten bykse 1 yap (predict bize sigmoid den dolay proba degerleri dndrd.)
#y_pred = model.predict_classes(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### with class_weigth

Investigate how the "class_weight" hyper-parameter is used in a Neural Network.

n_samples=  43400,  n_classes= 2(0&1), n_sample0= 42617, n_samples1= 783
wj=n_samples / (n_classes * n_samplesj)
class_weights = { 0 : 1.2 , 1 : 0.9 }

```python
def class_weight(df):
    class_weight = {}
    class_weight[0] = df.shape[0] / (df["Exited"].nunique() * df[df["Exited"]==0].shape[0])
    class_weight[1] = df.shape[0] / (df["Exited"].nunique() * df[df["Exited"]==1].shape[0])
    return class_weight
class_weight(df)
```

## GridSearchCV

```python
from tensorflow.keras.optimizers import Adam, Adadelta, RMSprop, Nadam
```

```python
def build_classifier(optimizer, learn_rate):
    classifier = Sequential()
    classifier.add(Dense(units = 32, activation = 'relu'))
    classifier.add(Dense(units = 16, activation = 'relu'))
    classifier.add(Dense(units = 1, activation = 'sigmoid'))
    classifier.compile(optimizer = optimizer(learn_rate), loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
early_stop = EarlyStopping(monitor = "val_accuracy", mode = "auto", verbose = 1, patience = 15)
```

```python
tf.random.set_seed(seed)

classifier_model = KerasClassifier(build_fn = build_classifier, validation_split = 0.1, epochs = 200)

parameters = {'batch_size': [32, 64],
              'optimizer': [Adam, RMSprop, Adadelta, Nadam],
              'learn_rate': [0.001, 0.003, 0.005]}

grid_model = GridSearchCV(estimator = classifier_model,
                          param_grid = parameters,
                          scoring = 'accuracy',
                          cv = 10,
                          n_jobs = -1,
                          verbose = 1)

grid_model.fit(X_train, y_train, callbacks = [early_stop])
```

#### Evaluate

## Final Model and Model Deployment

```python
tf.random.set_seed(seed) 

final_model = Sequential()
final_model.add(Dense(32, activation = "relu", input_dim = X_train.shape[1]))   # sigunt ve hiperb. e gre renme daha fazla olduu iin relu kullandk.
final_model.add(Dense(16, activation = "relu"))
final_model.add(Dense(1, activation = "sigmoid"))  # output binary olduu (iin 1neuron var) sigmoid kullandk. 
opt = Adam(lr = 0.001)
final_model.compile(optimizer = "rmsprop", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
early_stop = EarlyStopping(monitor = "val_loss", mode = "auto", verbose = 1, patience = 25)
```

```python
final_model.fit(x = X_train, y = y_train, validation_split = 0.1, batch_size = 32, epochs = 1000, verbose = 1,
          callbacks = [early_stop], class_weight=class_weight(df))
```
